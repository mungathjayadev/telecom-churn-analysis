```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandasql import sqldf # Import pandasql for running SQL queries

# Configure visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.tight_layout()

# 2. Load the dataset
df = pd.read_csv("telecom_data.csv")
```

```python
# 3. Handle missing values/convert data types

# 'TotalCharges' is read as an object type due to ' ' (empty strings).
# Replace empty strings with NaN, convert to numeric, and fill NaN (where tenure=0) with 0.
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Create a numeric column for Churn (1 = Yes, 0 = No) for calculations
df['Churn_Numeric'] = df['Churn'].map({'Yes': 1, 'No': 0})

print("Data Cleaning Complete. Info after cleaning:")
df.info()
```

```python
# Initialize pandasql function
pysqldf = lambda q: sqldf(q, globals())

# 4. SQL Query 1: Overall Churn Rate & Customer Count by Internet Service

query_1 = """
SELECT
    InternetService,
    COUNT(customerID) AS TotalCustomers,
    SUM(Churn_Numeric) AS TotalChurned,
    ROUND(CAST(SUM(Churn_Numeric) AS REAL) * 100 / COUNT(customerID), 2) AS ChurnRate_Percent
FROM
    df
GROUP BY
    InternetService
ORDER BY
    ChurnRate_Percent DESC;
"""
print("--- SQL Query 1: Churn by Internet Service ---")
print(pysqldf(query_1))


# 5. SQL Query 2: Churn Rate by Contract Type

query_2 = """
SELECT
    Contract,
    ROUND(CAST(SUM(Churn_Numeric) AS REAL) * 100 / COUNT(customerID), 2) AS ChurnRate_Percent
FROM
    df
GROUP BY
    Contract
ORDER BY
    ChurnRate_Percent DESC;
"""
print("\n--- SQL Query 2: Churn by Contract Type ---")
print(pysqldf(query_2))


# 6. SQL Query 3: Average Monthly and Total Charges by Churn Status

query_3 = """
SELECT
    Churn,
    ROUND(AVG(MonthlyCharges), 2) AS AvgMonthlyCharge,
    ROUND(AVG(TotalCharges), 2) AS AvgTotalCharge
FROM
    df
GROUP BY
    Churn;
"""
print("\n--- SQL Query 3: Average Charges by Churn Status ---")
print(pysqldf(query_3))
```

```python
# 7. Visualization 1: Overall Churn Distribution (Pie Chart)

churn_counts = df['Churn'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(
    churn_counts,
    labels=churn_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'salmon'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Overall Customer Churn Distribution', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
# 8. Visualization 2: Churn Rate by Contract Type

churn_contract = df.groupby('Contract')['Churn_Numeric'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(8, 5))
sns.barplot(
    x=churn_contract.index,
    y=churn_contract.values,
    palette='viridis'
)
plt.title('Churn Rate by Contract Type', fontsize=14)
plt.ylabel('Churn Rate (%)')
plt.xlabel('Contract Type')
plt.ylim(0, 50)
plt.show()
```

```python
# 9. Visualization 3: Monthly Charges Distribution (Histogram)

plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    x='MonthlyCharges',
    hue='Churn',
    multiple="stack",
    kde=True,
    palette=['skyblue', 'salmon'],
    bins=30
)
plt.title('Monthly Charges Distribution by Churn Status', fontsize=14)
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Customer Count')
plt.show()
```

```python
# 10. Visualization 4: Churn Rate by Tenure Group

# Create tenure groups (bins)
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['0-12 Mo', '13-24 Mo', '25-36 Mo', '37-48 Mo', '49-60 Mo', '61-72 Mo']
df['Tenure_Group'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=True, include_lowest=True)

# Calculate churn rate by tenure group
churn_tenure = df.groupby('Tenure_Group')['Churn_Numeric'].mean().sort_index() * 100

plt.figure(figsize=(10, 5))
sns.barplot(
    x=churn_tenure.index,
    y=churn_tenure.values,
    palette='rocket'
)
plt.title('Churn Rate by Customer Tenure Group', fontsize=14)
plt.ylabel('Churn Rate (%)')
plt.xlabel('Tenure Group (Months)')
plt.show()
```
